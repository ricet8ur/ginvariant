```python
from scipy.linalg import schur
from matplotlib.axes import Axes

number_of_subplots = 6


def generate_example():
    global number_of_subplots
    G = np.zeros((3, 3))
    a, b = np.random.random(2) + 0.3
    d = max([a, b]) - min([a, b])
    c = np.random.random() * (a + b - d) + d
    #     *
    #   b |\ c
    #     | \
    #     *--*
    # phi   a
    #
    # [[0, *, *],
    #  [a, 0, *],
    #  [b, c, 0]]
    #
    G[1, 0] = a
    G[2, 0] = b
    G[2, 1] = c
    G = G + G.T

    # print(G)
    q = 0
    figsize = (15, 15)
    plt.figure(figsize=figsize)
    colors = [
        (0.9, 0, 0),
        (0, 0.9, 0),
        (0, 0, 0.9),
    ]

    def plot(G: np.ndarray):
        global number_of_subplots
        nonlocal q, figsize, colors
        q += 1
        a = G[1, 0]
        b = G[2, 0]
        c = G[2, 1]
        # sample_direction = (1, 0)
        phi = np.arccos((-(c**2) + a**2 + b**2) / 2 / b / a)
        C = (0, 0)
        B = (a, 0)
        A = (b * np.cos(phi), b * np.sin(phi))
        x = [C[0], B[0], A[0], C[0]]
        y = [C[1], B[1], A[1], C[1]]
        plt.subplot(1, number_of_subplots, q)
        plt.scatter(x, y, s=50, c=colors + [colors[0]])
        plt.plot(x, y, color=(0, 0, 0))
        plt.axis("square")
        plt.xlim([-1.4, 1.4])
        plt.ylim([-1.4, 1.4])

    def permute(G: np.ndarray):
        nonlocal colors
        colors = colors[1:] + [colors[0]]
        K, V = schur(G)
        # print(f"{K=}, {V=}")
        print(f"permute start: \n\t{V=},\n\t{V @ K @ V.T=}")
        # print(f"{G=}, {V @ K @ V.T=}")
        W = V.copy()
        W[0, :] = V[1, :]
        W[1, :] = V[2, :]
        W[2, :] = V[0, :]
        # print(f"{V=}, {D=}")
        print(f"permute end: \n\t{W=},\n\t{W @ K @ W.T=}")
        # print(f"{G=}, {D @ K @ D.T=}")
        return W @ K @ W.T

    def transpose(G: np.ndarray, a: int, b: int):
        nonlocal colors
        colors_copy = list(colors.copy())
        colors[b] = colors_copy[a]
        colors[a] = colors_copy[b]
        K, V = schur(G)
        # print(f"{K=}, {V=}")
        print(f"transpose({a=} {b=}) start: \n\t{V=},\n\t{V @ K @ V.T=}")
        W = V.copy()
        W[b, :] = V[a, :]
        W[a, :] = V[b, :]
        # print(f"{V=}, {W=}")
        print(f"transpose({a=} {b=}) end: \n\t{W=},\n\t{W @ K @ W.T=}")
        return W @ K @ W.T

    # def inverse(G: np.ndarray, a: int):
    #     K, V = schur(G)
    #     # print(f"{K=}, {V=}")
    #     # print(f"{G=}, {V @ K @ V.T=}")
    #     V[:, a] = -V[:, a]
    #     # print(f"{V=}, {D=}")
    #     # print(f"{G=}, {V @ K @ V.T=}")
    #     return V @ K @ V.T

    # def inverse2(G: np.ndarray, a: int):
    #     K, V = schur(G)
    #     # print(f"{K=}, {V=}")
    #     # print(f"{G=}, {V @ K @ V.T=}")
    #     V[a, :] = -V[a, :]
    #     # print(f"{V=}, {D=}")
    #     # print(f"{G=}, {V @ K @ V.T=}")
    #     return V @ K @ V.T

    K, V = schur(G)
    print(f"{V=}")
    print(f"{G=}")
    plot(G)
    print(f"{G=}")
    G = permute(G)
    print(f"{G=}")
    plot(G)
    G = permute(G)
    print(f"{G=}")
    plot(G)

    G = transpose(G, 1, 0)

    G = permute(G)
    print(f"{G=}")
    plot(G)
    G = permute(G)
    print(f"{G=}")
    plot(G)
    G = permute(G)
    print(f"{G=}")
    plot(G)


generate_example()
plt.title('example of all physical rotations')
```
